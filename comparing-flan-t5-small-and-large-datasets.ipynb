# FLan-T5
### **F**ine-tuned **LA**nguage **N**et -- **T**ext-**T**o-**T**ext **T**ransfer **T**ransformer
This Google Research Large Language Model (LLM) is an encoder-decoder model pre-trained on a mixture of tasks. Each task is converted into a text-to-text format.

This code compares two datasets of FLan-T5, small and large.

```python
!pip install transformers
```

Input your own prompt and see the result in FLAN-T5 small dataset (300M).

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

#Input your own prompt below:
inputs = tokenizer("Explain artificial intelligence", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

Use the following prompts or input your own to compare FLAN-T5's large dataset (3G) to the previous small dataset output. 

```python
my_text = 'Explain artificial intelligence'
#my_text = 'What is ChatGPT?'
#my_text = 'A short introduction to machine learning:'
```

```python
%%time

inputs = tokenizer(my_text, return_tensors="pt")
outputs = model.generate(**inputs, \
                         min_length=256, \
                         max_new_tokens=512, \
                         length_penalty=2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         early_stopping=True)

output_text_Flan_t5 = tokenizer.batch_decode(outputs, \
                          skip_special_tokens=True)

print(output_text_Flan_t5)
```
